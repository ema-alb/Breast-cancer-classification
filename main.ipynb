### Local check
---

The first thing to do is to check if data is already present inside the working directory. With `check_local_files()` we perform this and, in case no files are found (more specifically the script looks for the directory *./data_folder*) it will perform the `retrieve_data()` function, otherwise it just notifies that X files have been found out of the Y that are actually available on the website.

```python
%matplotlib inline

import requests
import json
import re
import os
import gzip
import shutil
import tarfile
import warnings

# Suppress Warnings
warnings.filterwarnings("ignore")

DATA_FOLDER_NAME = "data_folder"


def retrieve_data():
    print("Retrieving data from server: gdc.cancer.gov")
    get_files_from_uuid_list(get_uuid_list())

def check_local_files():
    print("Performing check...")
    data_folder = os.path.join(os.curdir, DATA_FOLDER_NAME)
    # check if the data directory exists
    if not os.path.isdir(data_folder) or not os.path.exists(data_folder):
        print("No samples have been found! Downloading them now...\n")
        retrieve_data()
    else:
        n_uuid_files_local = len(os.listdir(data_folder))
        n_uuid_files_online = len(get_uuid_list())
        if n_uuid_files_local != n_uuid_files_online:
            print("The local collection might not be up to date...\n{}/{} samples available".format(n_uuid_files_local, n_uuid_files_online))
        else:
            print("There are {} samples available locally.".format(n_uuid_files_local))
    print("Check completed.")
```

### Getting UUIDs
---

Now we have to get the list of the UUIDs (the identifier of the files) of the files that respect the filters that we want to set, so we create a function called `get_uuid_list()` where we retrieve all the UUIDs and the names of the files that we will download.

Each of the files is a sample containing all the genome of a person.
The filters to be set are inside the `filters` object.

The returned value is a dictionary where:
- `key` = UUID, identifying the file to be downloaded (for example 'd094f341-9b5b-4f34-8321-f6d97d31a9b3')
- `value` = file name, which is the Case ID of the sample (for example 'TCGA-5L-AAT1')

```python
def get_uuid_list():
    files_endpt = "https://api.gdc.cancer.gov/files"

    filters = {
        "op": "and",
        "content": [
            {
                "op": "in",
                "content": {
                    "field": "cases.project.primary_site",
                    "value": ["Breast"]
                }
            },
            {
                "op": "in",
                "content": {
                    "field": "cases.project.project_id",
                    "value": ["TCGA-BRCA"]
                }
            },
            {
                "op": "in",
                "content": {
                    "field": "files.analysis.workflow_type",
                    "value": ["HTSeq - FPKM-UQ"]
                }
            },
            {
                "op": "in",
                "content": {
                    "field": "files.cases.samples.sample_type",
                    "value": ["Primary Tumor"]
                }
            }
        ]
    }

    # Here a GET is used, so the filter parameters should be passed as a JSON string.
    params = {
        "filters": json.dumps(filters),
        "fields": "file_id",
        "fields": "associated_entities.entity_submitter_id",
        "format": "JSON",
        "size": "2000"  # arbitrary value
    }

    # "fields": "associated_entities.entity_submitter_id"
    response = requests.get(files_endpt, params=params)

    # This step populates the download list with the file_ids from the previous query
    uuid_tgcaName = {}  # dictionary : {'uuid': Name of the tcga patient}
    for file_entry in json.loads(response.content.decode("utf-8"))["data"]["hits"]:
        # take the first 12 chars of the entity_submitter_id. The [0] is because we have "associated_entities" that can have multiple values inside of it (entity_id, etc...)
        tgca_name = file_entry["associated_entities"][0]["entity_submitter_id"][:12]
        file_uuid = file_entry["id"]
        uuid_tgcaName[file_uuid] = tgca_name

    return uuid_tgcaName
```

### Download phase
---

With the dictionary that we receive from the `get_uuid_list()` function we can go ahead and download all the files using the method `get_files_from_uuid_list(uuid_tgcaName)`.

To do so we do a post to the data endpoint `data_endpt = "https://api.gdc.cancer.gov/data"` passing the list of the keys of the dictionary (with all the UUIDs to be downloaded).

The result of the download will be a big zipped file that will contain *N* archives, where N is the number of samples that are available. In order to easily deal with all the samples files, we have to extract this archive, along with each of the archives inside of it. The function `extract_tar_gz(archive_name, data_folder, uuid_tgcaName)` does this, receiving the following parameters:
- *archive_name*, which is the name of the single big archive, containing all the others
- *uuid_tgcaName*, the dictionary with key = UUID and value = Case ID

At the end of this method we will be left with a *data_folder* directory created inside the working directory, with all the samples saved as files with name equal to the Case ID.

```python
def get_files_from_uuid_list(uuid_tgcaName):
    data_endpt = "https://api.gdc.cancer.gov/data"
    uuids = list(uuid_tgcaName.keys())  # retrieve only the uuids that I will download
    params = {"ids": uuids}
    response = requests.post(data_endpt, data=json.dumps(params), headers={"Content-Type": "application/json"})
    response_head_cd = response.headers["Content-Disposition"]
    file_name = re.findall("filename=(.+)", response_head_cd)[0]

    # create the first part of the path, starting from the current dir
    data_folder = os.path.join(os.curdir, DATA_FOLDER_NAME)
    file_name = os.path.join(data_folder, file_name)
    # create dir: if already existing, exist_ok=true suppresses any errors
    os.makedirs(os.path.dirname(file_name), exist_ok=True)
    with open(file_name, "wb") as output_file:
        output_file.write(response.content)
        print("Filename: {}".format(file_name))

    extract_tar_gz(file_name, uuid_tgcaName)


def extract_tar_gz(archive_name, uuid_tgcaName):
    with tarfile.open(archive_name, "r") as ap:
        # ap is the archive pointer. For each directory encountered extract it and analyze it
        for element in ap.getmembers():
            if element.name != "MANIFEST.txt":
                # extract dir
                ap.extract(element.name, data_folder)
                file_gz_path = os.path.join(data_folder, os.path.join(element.name.split("/")[0], element.name.split("/")[1]))
                # open and extract the last archive, that contains only the txt file we're interested in
                uuid_file = element.name.split("/")[0]
                # name of the output file that will contain the genome of that specific TCGA patient
                file_txt_path = os.path.join(data_folder, uuid_tgcaName[uuid_file])
                try:
                    # extact the .gz file using the gzip library, opening two fp and copying the content of the 1st into the 2nd
                    with gzip.open(file_gz_path, 'rb') as f_in, open(file_txt_path, 'wb') as f_out:
                        shutil.copyfileobj(f_in, f_out)
                except Exception as e:
                    print("Generated exception: {}".format(type(e).__name__))
                # remove the folder that has just been extracted, because now it's useless
                shutil.rmtree(os.path.dirname(file_gz_path))

    # after having extracted the .tar.gz file, remove it
    os.remove(archive_name)
```

### Importing data into memory
---

After having downloaded all the necessary sample files, it's time to transfer the data into memory via the function `get_data_from_files(only_labelled=False)`.

The returned values are:
- `M`, the matrix containing the samples which can be only the labelled ones, if the `only_labelled` flag is set to , or also the unlabeled ones otherwise
- `y`, the list containing all the labels
- `ensembl_gene_list`, the list containing all the names of the genes in the Ensembl nomenclature
- `label_mapping`, the list containing the map between numerical and categorical label

The idea is to create a $NxM$ matrix `M` where:
- $N$ = # of samples
- $M$ = # of features (genes)

This matrix will contain only the samples, whereas the `y` list will contain the labels and its size is equal to $N$, equal to the # of samples.
In this way we for the sample $M_i$ the corrisponding label will be $y_i$.

The program dinamically maps the corrispondance between the value of $y_i$ and the nominal value of the label, so we need a list containing the mapping. The `label_mapping` list does this and it's structured in this way:
- `index` = numerical label, in the range of $[0, 4]$
- `value` = categorical label, which is the actual subtype of breast cancer. Possible values: *Luminal A, Luminal B, Basal-like, HER2-Enriched, Normal-like*

Another thing that is extracted from the files is the names in the Ensembl terminology of each gene. The `ensembl_gene_list` is structured this way:
- `key` = index of the gene inside of a sample file
- `value` = name of the gene (for example '*ENSG00000242268.2*')

```python
import numpy as np

DATA_FOLDER_NAME = "data_folder"

def get_data_from_files(only_labeled=False):
    if only_labeled:
        DATA_FOLDER_NAME = "data_folder_labelled"
    else:
        DATA_FOLDER_NAME = "data_folder"
    
    data_folder = os.path.join(os.curdir, DATA_FOLDER_NAME)
    ensembl_gene_list = []
    M = []
    y = []
    # "y" is the list containing the labels of the matrix of the data
    #label_mapping = }{ # ["Luminal A"] = 0, ["Luminal B"] = 1, etc...
    label_mapping = []  # list: [0] = "Lumi nal A", [1] = "Luminal B", etc...
    id_label = build_associative_array_caseID_label()   # dictionary: {"TCGA-AAAA": "Luminal A"}
    i = 0
    label_counter = 0
    for filename in os.listdir(data_folder):
        # check if the data we are analyzing is labeled, if not, skip it
        if(only_labeled):
            if filename in id_label.keys():
                # check if we already encoutered a certain label: if not, add it to the collection of labels
                if id_label[filename] not in label_mapping:
                    #label_mapping[id_label[filename]] = label_counter
                    label_mapping.append(id_label[filename])
                    label_counter += 1
                # add the label of the i-th element inside the Y list, containing all the labels of each sample
                y.append(label_mapping.index(id_label[filename]))
                filename = os.path.join(data_folder, filename)
                with open(filename, "r") as fp:
                    j = 0
                    gene_data = []
                    line = fp.readline()
                    while line:
                        gene_data.append(float(line.split("\t")[1]))
                        if i==0:
                            ensembl_gene_list.append(line.split("\t")[0])
                        line = fp.readline()
                        j += 1
                    M.append(np.array(gene_data))
                    i += 1
        else:
            # do this when we want all the data, even the ones not labelled
            filename = os.path.join(data_folder, filename)
            with open(filename, "r") as fp:
                j = 0
                gene_data = []
                line = fp.readline()
                while line:
                    gene_data.append(float(line.split("\t")[1]))
                    if i==0:
                        ensembl_gene_list.append(line.split("\t")[0])
                    line = fp.readline()
                    j += 1
                M.append(np.array(gene_data))
                i += 1

    # script per creare un file contenente solo
    # with open("gene_names.txt", "w") as fout:
    #     for ens in ensembl_gene_list:
    #         str = get_gene_name(ens)+'\n'
    #         fout.write(str)

    gene_list = []
    with open("gene_names.txt", "r") as fp:
        i = 0
        line = fp.readline()
        while line:
            if line != '---':
                gene_list.append(line.strip('\n'))
            else:
                gene_list.append(ensembl_gene_list[i])
            line = fp.readline()
            i += 1

    return M, gene_list, y, label_mapping
```

### Build Associative Array
---

This step is about building a data structure to associate every sample to a label. The whole `label_file` is read line by line and, if a label is given, a new entry *{CaseID, Label}* is added to the array. At the end the `label_array` is returned.

```python
INPUT_LABEL_DATA = "input.txt"

def build_associative_array_caseID_label():
    label_file = os.path.join(os.curdir, INPUT_LABEL_DATA)
    label_array = {}
    i=0
    with open(label_file, "r") as fp:
        while True:
            line = fp.readline()
            if not line:
                break
            else:
                label=line.split("\t")[20]
                if label=="NA":
                    break
                CaseID = line.split("\t")[0]
                label_array[CaseID] = label
                
    return label_array
```

### Convert Ensembl to HGNC Symbol nomenclature
---

The `get_gene_name(ens_name)` function is in charge of translating a given `ens_name` Ensembl gene into the corrisponding name in the HGNC (HUGO Gene Nomenclature Committee) Symbol nomenclature, using the dictionary file `gene_dictionary_sorted.txt` which downloaded from the [Ensembl website](https://www.ensembl.org/ "Ensembl official website")

What the script does is, once received the parameter, search inside each line of the file and return the HGNC Symbol name right when the corrispondance with the Ensembl name is found.

The format of the file is the following 
>Ensembl name,HGNC Symbol name

```python
DICTIONARY_FILE = "gene_dictionary_sorted.txt"

def get_gene_name(ens_name):
    dict_path = os.path.join(os.curdir, DICTIONARY_FILE)
    ens_name = ens_name.split('.')[0]
    result = ''
    with open(dict_path, 'r') as f:
        for line in f:
            if line.startswith(ens_name):
                return line.split(',')[1]
    return result
```

### Possible solutions
---

*DISCLAIMER*: In our study case we decided to deal also with the "bad" class *Normal-like*, even if it had very few samples. What we know is that our classifiers might not be very good at classifying this class since we have not enough test data to check the goodness of our model. This is also the reason why in the following description of the balancing method some decisions were made. All of them will be detailed later on.

There's a need to balance the data, so approaches that can be taken are four. The first three involve what is called "*resampling*" that involves the elimination or the addition of samples to the dataset, whereas the last one is the most naive one:
1. **Undersampling**
2. **Oversampling**
3. **Mixed approach (over+under sampling)**
4. **Partially ignore the problem**

Let us examine how each of these techniques work and how we can apply them, if possible, to our case. 

### 2.1 Undersampling
---

This technique removes samples from the original dataset, so that the resulting one will have a smaller cardinality.

![title](https://i.imgur.com/7ynhdIx.png)

There are several different algorithms to do so. We can divide them in two main categories: *controlled under-sampling techniques*, which allow to determine how many samples for each class we want to be left with, and *cleaning under-sampling techniques*, which doesn't permit that.

Some algorithms are:
- **Random undersampling**, in which we under-sample the majority class(es) by randomly picking samples with or without replacement. This is a *controlled under-sampling technique*;
- **Edited Nearest Neighbors**, in which we apply a Nearest-Neighbors algorithm and "edit" the dataset by removing samples which do not agree "enough" with their neighboorhood. This is a *cleaning under-sampling technique*;
- **Tomek's Link**, in which we aim to remove the Tomek's links, that will be described in detail later.

Since in our data set the least represented class is *Normal-like* with 8 samples, balancing with either of these methods would cause a huge loss in information. In fact, we would have to balance all the classes reducing their cardinality to 8 samples each, being left with a total dataset of 40 samples: definitely a waste of what we had!

We could even remove the class *Normal-like*, but it was our intention to keep considering it, so for these reasons the undersampling approach is retained to be a **bad solution** for our case.


### 2.2 Oversampling
---

This technique adds samples to the original dataset, so that the resulting one will have a bigger cardinality.

![Imgur](https://i.imgur.com/wXVfaa9.png)

As before, there are more naive and less intuitive methods to achieve the goal. The main algorithms to perform this are:
- **Random oversampling**, similarly to the undersampling version, we over-sample the minority class(es) by randomly picking samples with or without replacement.



- **SMOTE (Synthetic Minority Oversampling TEchnique)**, we create new elements for the minority class(es), based on those that already exist. It works by randomly picking a point from the minority class and computing the k-nearest neighbors for this point. The synthetic points are added between the chosen point and its neighbors.
![Imgur](https://i.imgur.com/s2Gt2Lq.png)



- **ADASYN (ADAptive SYNthetic)**, which works like SMOTE, but it adds a random small values to the points thus making it more realistic, so that the randomly generated samples look more scattered than SMOTE.
<img src="https://imgur.com/KM6RzC5.png" alt="SMOTE vs ADASYN" style="max-width:80%; max-height:80%;"/>



For our study case we decided to choose the SMOTE algorithm with the SVM variant (**SVMSMOTE**), which is an extension that uses an SVM classifier to find support vectors and generate samples considering them. Since the SVC has some parameters (such as *C, gamma* and *kernel*), we will set the best ones found with the grid search detailed in the *Classification* part. By using this algorithm we are sure that we create new samples inside a region delimited by the Support Vectors, as we can see from the picture below.
<img src="https://imgur.com/ylDRh9o.png" alt="SVMSMOTE" style="max-width:75%; max-height:75%;"/>


### 2.3 Mixed approach (over+under sampling)
---

This approach applies oversampling first and then undersampling. The reason why we would want to do both, in this specific order, is that SMOTE might generate noisy samples by interpolating new points between marginal outliers and inliers, so we need some cleaning of the newly created samples to have a better defined dataset. 

We always start with **SMOTE** for the oversampling part, then we can choose between **Edited Nearest-Neighbours** or **Tomek's Link**:
- **SMOTEENN**, tends to remove many noisy samples; this approach is not good in our case since, as we will see in the *Clustering* section, we noticed that the data is very sparse, thus we risk to remove too many real samples, giving too much importance to the artificially created one, also losing much information.
- **SMOTETomek**, which removes the *Tomek's links*. These are pairs of very close instances, but of opposite classes. Removing these couples increases the space between two nearby classes, facilitating the classification process.
![title](https://imgur.com/rL5mz8b.png)


Below we can se the difference between applying SMOTE only (left picture), SMOTE + ENN (center picture), SMOTE + Tomek (right picture).

![title](https://imgur.com/co8Lm64.png)
As we can observe, the SMOTE creates many noisy points. The SMOTE + ENN tends to clean a lot, but, as said earlier, it might clean too much, resulting in an important information loss. The last one (**SMOTETomek**) is the best tradeoff in our opinion, since it removes only those points that might create the most confusion, that is the ones lying close to the borders between different classes.


### 2.4 Partially ignore the problem
---

Even though at first this might seem an absolute failure, we wanted to scientifically prove whether this approach is totally bad for our case or, for this specific case, it might be a possible alternative.

The premise is that, since this approach aims at leaving a certain class imbalance, we will not look at *accuracy* only when evaluating the performances of our model, since that might be hugely misleading, but we will also look at the *F1-score*, which takes into account both the *precision* and the *recall* scores. As a matter of fact, this indicator comes in really handy with problems dealing with imbalanced classes.

Why are we talking about *partially* ignoring the problem? The reason behind this is that we are going to leave the class imbalance, but we are making sure that this imbalance is not as bad as it actually is right now. As discovered earlier, the *Luminal A* class has an advantage towards *Normal-like* of 230:8, that is the latter is just ~3.5% of the former! This is due to the very low cardinality of the *Normal-like*, that can be addressed by an expansion of the minority class to reach a certain proportion towards the majority class. This value can be set by the user, but we set it to 25%. In this way the number of samples for the *Normal-like* class will become 57, adding 49 samples to the initial 8. After performing this, we increase the samples for all the other classes (excluding the majority class) of the same number, having the majority class as upper bound: in this way all the classes gain importance with respect to the majority class, but keeping the imbalance, even if less marked.

This is what the programmer called *proportional oversampling*.

The decision for this was due to the observation of real life samples, that tend to be *Luminal A* rather than *HER2-enriched* for example. By explicitly keeping the bias, but imposing a minimum number of samples for the minority class, we want to verify if this leads to a bad result at classifying even the minority classes.

Here is the algorithm that perform the proportional oversampling.

```python
def oversample_proportionally (X, y, os_algorithm, max_min_ratio=0.25):
    occurrances = Counter(y).most_common()  # create a dictionary {label: occurrance} ordered from the most present to the least present
    dict = {}

    # ---- increase the number of samples of the minority class and save the percentage increase
    initial_ratio = occurrances[len(occurrances)-1][1]/occurrances[0][1]
    if initial_ratio < max_min_ratio:
        incremental_ratio = max_min_ratio - initial_ratio
        dict[occurrances[len(occurrances)-1][0]] = int(occurrances[0][1] * max_min_ratio)

        i = 0
        for x in occurrances:
            # do not do this for the maj and min class
            if i != len(occurrances)-1 and i != 0:
                # calculate the samples to be added
                new_samples = int(occurrances[0][1] * incremental_ratio)
                # the number of samples of the maj class is the upper bound
                if x[1] + new_samples < occurrances[0][1]:
                    dict[x[0]] = x[1] + new_samples
                else:
                    dict[x[0]] = occurrances[0][1]
            i += 1

        X, y = os_algorithm.fit_resample(X, y)

    return X, y
```

As mentioned earlier, the scores on which we will base our evaluations are, for each class:
- **Accuracy**, is the ratio of correctly predicted observation to the total observations. $$Accuracy = \frac{TP+TN}{TN+TP+FN+FP} = \frac{\#\,correct\,predictions}{\#\,all\,samples}$$
- **Recall**, is the ratio of relevant instances that have been retrieved over the total amount of relevant instances. More simply, how many of the Actual Positives our model capture through labeling it as Positive (True Positive). $$Recall = \frac{TP}{TP+FN} = \frac{\#\,correct\,predictions}{\#\,all\,possible\,correct\,istances}$$
- **Precision**, is the ratio of relevant instances that have been retrieved over the total amount of relevant instances. More simply, how precise your model is out of those predicted positive, how many of them are actual positive. $$Precision = \frac{TP}{TP+FP} = \frac{\#\,correct\,predictions}{\#\,all\,predictions}$$
where: 
$$TP\,=\,True\,Positive\qquad FP\,=\,False\,Positive\qquad TN\,=\,True\,Negative\qquad FN\,=\,False\,Negative$$
- **F1 Score**, is an harmonic mean of Precision and Recall. $$F1\,Score = \frac{2\times Precision\times Recall}{Precision+Recall}$$


*Accuracy* is very good to evaluate the performances of balanced datasets, but with imbalanced datasets it could be very misleading. This is why whenever we will deal with imbalanced test sets, we will pay more attention to the *F1 Score*, because it gives us more useful indication on how the model performs. Also, in case errors have different weights, this is the measure to look at.

### Feature Selection - Step #0 : filter approach

As a first step we have discarded features that have a value of variance equal to 0 or equal to a value that is very close to zero. We considered those features as useless because they don't bring any significant information that helps in the classification and, in addition to that, they slow down the program.

### Option #2 : PCA-based Feature Selection

We exploited PCA to develope a custom approach to perform Feature Selection.

Since each principal component can be expressed by a linear recombination of the starting features, we have thought to take advantage of the weights assumed by each feature in order to identify the ones that contribute the most: the higher the value of the weight/coefficient (considering its absolute value) the more important the feature in determining that PC. Looking at the grid below, we can see that each column shows the set of weights assumed by each feature in the expression of all the 10 Principal components considered.

```python
print('Weights of features for each principal component:')
df = pd.DataFrame(pca.components_[:num,:]*10000)
df.index.name = 'PC'
df.columns.name = 'Features'
pd.set_option("display.precision",6)
df
```

#### Method Explanation:
More in detail we create a ranking of all the features based on a value of importance that represents how much each feature is influential in determining the first 10 principal components. Considering one feature at a time (i.e. one column at a time), we summed all its contributes weighted by the explained variance of each PC. In other words we calculate a value that reveals how much each feature has had a determinant role in estimating the 10 most important Principal Components.

Starting from the plot of the feature importances, we performed a search for the best threshold value to apply in order to reject features below that fixed limit.

```python
count = 0
feature_importance = []
useful_columns_RF = []
num_pc = pca.components_.shape[0]
num_features = pca.components_.shape[1]

for j in range(num_features):  # iterate by columns/features
    sum = 0
    for i in range(num):
        sum = sum + abs(pca.components_[i, j]) * (pca.explained_variance_ratio_[i])    #  Weighted sum of absolute value of feature contribution for all PCs

    feature_importance.append(sum)

importances = feature_importance

# Plot histogram of feature importances
plt.figure()
plt.title("\nFeature Importance - PCA")
plt.yticks(np.arange(min(importances), max(importances), 0.05))
plt.grid(True)
plt.plot(importances)

#Set Threshold 0.006633
th = 0.012 * np.max(feature_importance)  # fix a threshold for selecting features

i = 0
useful_columns = []

# Scan weighted sum and identify indexes which correspond to columns that has to be deleted
for val in feature_importance:
    if val >= th:
        count = count + 1
        useful_columns.append(i)
    i = i+1

print("FS PCA-based output:\n")
print(count, "features will be kept!")
print()

X_important = np.empty((filtered_gene_data.shape[0], len(useful_columns)))

k = 0
for i in range(num_features):
    if i in useful_columns: # If TRUE --> copy that column(feature), Else skip that column because it's not useful
        X_important[:, k] = filtered_gene_data[:, i]
        k = k+1

#Evaluate performances: Dataset's features filtered with PCA approach
acc, f1, prec, rec  = NB_Classifier(X_important, y, 1)
add_scores(acc, f1, prec, rec, "NB", 2)

acc, f1, prec, rec  = Support_Vector_Classifier(X_important, y, 1)
add_scores(acc, f1, prec, rec, "SVM", 2)

acc, f1, prec, rec = RandomForest_Classifier(X_important, y, 1)
add_scores(acc, f1, prec, rec, "RF", 2)
```

The key-point behind a random forest is to combine many decision trees into a single model to get more accurate predictions. In order to increase diversity in the forest and obtain more robust overall results, each decision tree can access only a subset of features and a subset of the training samples.

Since decision trees use features to split the data into subsets which most heavily belong to one class, we have thought to exploit the RandomForest Classifier to get a subset of features which are the most effective in determining these splits. Every node in each Decision Tree of the Forest is a condition on a single feature which is evaluated considering the level of impurity obtained downstream calculated by means of the GINI Index.

The reported plot shows the importance of every feature based on how much it decreases the impurity in a tree. In order to select only a subset of the features we have performed a search of the best threshold value, trying different values and selecting the best one as a trade off between the number of features selected and the performances obtained.

### Support Vector Classifier
---
The SVC aims at finding hyperplanes able to separate different classes. One hyperplane is identified by the vector $w = \displaystyle\sum_{i=1} y_iα_ix_i$ From this formula, we can immediately understand that the hyperplan is a linear combination of the training data $(x_i)$ weighed by the relative label $(y_i = {+1, -1})$ and by the $α$ coefficient. $α$ is different from zero only for the points located on the margins. These points (the support vectors) are the only points needed by the SVC to provide the classification. Since $α$ can assume values between 0 and C, C becomes the constant that determines how soft the margin we are using is.

<img src="https://imgur.com/zYWbfNe.png" width="550">  


Parameters:
- ***Kernel***
  
  Not every dataset is linerly separable. Often, the data are not easily separable with a linear hyperplane. In order to deal with this problem it is possible to extend the SVC with a non linear function, mapping the data into a new space with more dimensions. This requires a remarkable computational effort. To avoid this situation, it is possible to use the *kernel trick*. This trick consists of using a kernel function (we have considered four different kernels) which can ***implicitly*** transform the dataset into a higher-dimensional space with minimal effect on computation time.
  
  
- ***C***

  As we said before, C determines the tradeoff between misclassified data and width of the margin. As C is changing, the results will change. The lower C is, the wider the marging will be. The higher C is, the lower the number of misclassified data will be. What does this mean? It means that if we are using a low value of C the SVC will try to find the widthest marging possible, even if it means misclassifying some data. If we are using a high value of C, the SVC will try to reduce the number of misclassified data, even if it means reducing the width of the margin. This is the reason why the boundaries change.




- ***Gamma***
   
  This parameter is useful whenever non-linear kernel are used. This parameter changes a lot how the shape of the final decision boundary will be: with low values we have a wide gaussian distribution, leading to have model behaving similar to the linear kernel, whereas high values lead to narrow gaussian distribution, with a decision boundary that tends to have the shape of the data.

We have performed a ***grid search*** to find the best kernel, C and gamma, since if the number of features is much greater than the number of samples, choosing Kernel functions and regularization terms is crucial to avoid overfitting.

Firstly, we have split the dataset into test set (30%) and training set (70%). 

```python
X_train, X_test, y_train, y_test = train_test_split(gene_data, y, test_size=0.30, random_state=3)
```

After that, we have decided to create fictitious samples with SMOTETomek() only in the training set, leaving the test set as a *real* test set made only of real data. Then, we have performed a Stratified K Fold (with k=5) on the training set amplified.

```python
X_train, y_train = SMOTETomek().fit_resample(X_train, y_train)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
```

We have selected a range of values for the parameter C: {0.001, 0.01, 0.1, 1, 10, 100, 1000}, gamma: {0.001, 0.01, 0.1, 1, 10, 100, 1000} and kernel: {'linear', 'poly', 'rbf', 'sigmoid'}. At each iteration, we have created a model with the specified kernel, C and gamma and we have evaluated its accuracy on the five folds. After that, we have calculated the mean of these values of accuracy. Finally, we have chosen the combination of the values with the highest mean as the best parameters.

```python
from statistics import mean

c_range = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
gamma_range = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
kernel_rage = ['linear', 'poly', 'rbf', 'sigmoid']
nfold = 0
best_acc = -1

for kernel in kernel_rage:
    for c in c_range:
        for gamma in gamma_range:
            accuracy_list = []

            nl_SVC = SVC(kernel=kernel, C=c, gamma=gamma)

            for train_index, test_index in kf.split(X_train, y_train):
                x_train_fold, x_test_fold = np.asarray(X_train)[train_index], np.asarray(X_train)[test_index]
                y_train_fold, y_test_fold = np.asarray(y_train)[train_index], np.asarray(y_train)[test_index]

                nl_SVC.fit(x_train_fold, y_train_fold)

                predictions = nl_SVC.predict(x_test_fold)
                acc = accuracy_score(y_test_fold, predictions)
                accuracy_list.append(acc)

            print("SVC kernel ", kernel, " mean accuracy for C = ", c, "\tgamma\t= ", gamma, "\t: %3.2f%%"% (mean(accuracy_list)*100))

            if(mean(accuracy_list) > best_acc):
                best_acc = mean(accuracy_list)
                best_C = c
                best_gamma = gamma
                best_kernel = kernel
```

We have found that the best ***kernel*** is ***linear***, the best ***C*** is ***0.001*** and the best ***gamma*** is ***0.001***.

### Performance evaluation

```python
nl_SVC = SVC(kernel=best_kernel, C=best_C, gamma=best_gamma)
nl_SVC.fit(X_train, y_train)

predictions = nl_SVC.predict(X_test)
acc = accuracy_score(y_test, predictions)

print("SVC | Evaluation on test set:")
print("Accuracy calculated with kernel = ", best_kernel, " bestC = ", best_C, " bestGamma = ", best_gamma, " : ", acc)
```

The “forest” actually is an ensemble of Decision Trees, randomness is added to the model in order to increase the diversity and get more accurate predictions.
Features to be evaluated are randomly spread into the Forest and at every node in each Decision Tree the attribute on which we perform the split is the best locally to the random subset of features that has been chosen for that specific Decision Tree: that implies that the split is surely the best locally (with regards to the Decision Tree), but not absolute.

Creating random subsets of the features means building many smaller trees working on these subsets and combining them later on. A higher number of trees can increase the computation time but is a way to improve the performance and prevent overfitting: while the variance of the model decreases, the bias doesn’t change, resulting in more stable predictions.

Hyperparameters can be set to reduce the computational time or to increase the predictive power of the model.

### (Bonus) One vs. All Tree
---

This model comes from the application of two main models: **Decision Trees** and **One vs. All SVM**. Both these models have in common that they are made for **multiclass** problems.

The idea is to allow the data to be imbalanced, but classifying using more decision trees in cascade. In a decision tree we can, starting from the root node, reach a leaf that will label the data with one of the many labels available, that is we can solve multiclass problems. Instead of using a single tree to assign the label, we create $N-1$ trees, where $N = \#\,of\,classes$, each of which solves a binary classification problem. More specifically, we ask our model whether the sample belongs to class *Y* or not, and we keep iterating over the different models until one label for the data is found: this is the one-vs-all approach.

The order in which the models are run at classification time is not random: the first model will tell us whether the data belongs to the majority class or not. If not, the second model tells if the element belongs to the 2nd most represented label, and so on... We will end up with the last decision tree that will discrimine between *HER2-Enriched* and *Normal-like*.

The training of the models is made in a way that leaves an overall class imbalance, but in each single binary problem we have a balanced situation. In order to achieve this, we start training our models from the minority classes.

```python
from imblearn.over_sampling import SVMSMOTE
from imblearn.over_sampling import SMOTE
from sklearn.svm import SVC


def balance_for_50_50(X, y):
    counts = Counter(y).most_common()
    final_count = {}
    
    # least important class
    final_count[counts[-1][0]] = counts[-2][1]
    sum = final_count[counts[-1][0]]
    # we go from the least important to most important
    for i in range(-2, -len(counts)-1, -1):
        # counts[i][1] is the current value
        # if current value has smaller cardinality than all other classes
        if counts[i][1] < sum:
            # balance it
            final_count[counts[i][0]] = sum
        else:
            # leave it unchanged
            final_count[counts[i][0]] = counts[i][1]
        sum += final_count[counts[i][0]]

    #X, y = SVMSMOTE(svm_estimator=SVC(C=0.001, kernel='linear', gamma=0.001), sampling_strategy=final_count).fit_resample(X, y)
    X, y = SMOTE(sampling_strategy=final_count, random_state=3).fit_resample(X, y)
    
    return X, y
```

To evaluate the performances of the clustering algorithms, there are two main different type of numerical measures:
- **External Indexes**: evaluate the model taking advantage of the labels (extraneous to the context).
 - ***Homogeneity***: measures if each cluster contains only members of a single class.
 
 - ***Completeness***: measures if all members of a given class are assigned to the same cluster.
 
 - ***V-measure***: is the harmonic mean of homogeneity and completeness.


- **Internal Indexes**: evaluate the model without any external information.
 - ***Cluster cohesion***: measures how closely related are items in a cluster.
   - $Cohesion = \displaystyle\sum_{i}\displaystyle\sum_{x \in C_i} (x - \mu_i)^2 $ where $\mu_i$ is the mean of points in the set $C_i$

 - ***Cluster separation***: measure how distinct or wellseparated a cluster is from other clusters.
   - $Separation = \displaystyle\sum_{i}|C_i| (m - m_i)^2 $ where $m$ is the mean of points in one set and $|C_i|$ is the size of cluster $i$.
 
<img src="https://imgur.com/eVKhiX1.png" width="600"> 
 
 - ***Silhouette***: a succinct measure to evaluate how well each object lies within its cluster. It takes into account cohesion and separation.
   - $Silhouette = \frac{b-a}{max(a, b)} $ .

     - *a*: The mean distance between a sample and all other points in the same class.
     - *b*: The mean distance between a sample and all other points in the other cluster of which the sample is not a member.

### K-means

<img src="https://imgur.com/vJgL8PB.png">

K-means is a clustering algorithm which aims at creating globular clusters minimazing the total variance intra-cluster: $ arg min \displaystyle\sum_{i=1}^{K}\displaystyle\sum_{x_j \in S_i} ||x_j - \mu_i||^2 $ where $\mu_i$ is the mean of points in the set $S_i$. Each of these clusters is identified by a centroid or midpoint. The algorithm follows an iterative procedure. Initially, it selects K points as the initial centroids (this selections could be done randomly). Then, at each iteration, it calculates the distances between every point and the centroids and it assigns that point to the nearest cluster. At the end of each iteration, it calculates again the centroids and repeats the entire procedure. These steps are repeated until the centroids do not change anymore.

To understand which was the number of clusters to create in order to obtain the best clusterization possible, we have used the *SEE metric* in the Elbow Graph (Knee approach). The **Sum of Squared Error** is the sum of the squared differences between each item and its cluster's mean.

```python
from sklearn.cluster import KMeans
# Trying 10 k
K = range(1, 11)
SSE_list = []

for k in K:
    kmeansModel = KMeans(n_clusters=k, random_state=1)
    kmeansModel.fit(X)
    
    # Sum of squared distances of samples to their closest cluster center.
    SSE_list.append(kmeansModel.inertia_)
    
#Plot the elbow
plt.figure()
plt.plot(K, SSE_list)
plt.xlabel('K - number of centroids')
plt.ylabel('SSE')
plt.xticks(np.arange(1, len(K)), K)
plt.title('Elbow Graph showing the optimal k')
plt.show()
```

In this approch, we have computed for every model (created with a different value of k) the SSE and we have plotted the resulting graph. Since the K-means results could change according to the initial random centroids, we have run the algorithm many times and the picture has always shown a knee before k=6. From k=6 onwards, indeed, the SSE does not considerably decrease. This result is consistent with the knowledge we have a priori about the dataset. In fact, from the part concerning the classification, we know that there are only 5 labels.

### Hierarchical Clustering

<img src="https://imgur.com/IFzUhYb.png">

Hierarchical clustering is a method of cluster analysis which seeks to build a hierarchy of clusters. Strategies for hierarchical clustering generally fall into two types:
- **Agglomerative**
    
    This is a "bottom-up" approach: each observation starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy.
    
    
- **Divisive**

    This is a "top-down" approach: all observations start in one cluster, and splits are performed recursively as one moves down the hierarchy.
    
The algorithm follows an iterative procedure. Firstly, it computes the proximity matrix. After that, it let each data point be a cluster and than it starts the iterations. It keeps merging the two closest clusters and updating the proximity matrix until only a single cluster remains.

The results of hierarchical clustering are usually represented in a dendrogram, as we can see from the picture above.

In order to decide which clusters should be combined (for agglomerative), or where a cluster should be split (for divisive), a measure of **dissimilarity** between sets of observations is required. In most methods of hierarchical clustering, this is achieved by use of an appropriate metric (a measure of distance between observations), and a linkage criterion which specifies the dissimilarity of sets.
- The choice of an appropriate ***metric*** will influence the shape of the clusters, as some elements may be close to one another according to one distance and farther away according to another. In this analisys we focused on the main common metric: euclidean, l2, manhattan and cosine.

 - *$L^{2}$*: $||x||_2 = \sqrt{\displaystyle\sum_{i}^{n} a_i^2} = \sqrt{(a_1)^2 + (a_2)^2 + ... + (a_n)^2}$   

   - The Euclidean norm $L^{2}$ gives the ordinary distance from the origin to the point x.

 - *Euclidean*: $||a-b||_2 = \sqrt{\displaystyle\sum_{i}(a_i - b_i)^2}$

   - Euclidean metric is the "ordinary" straight-line distance between two points in Euclidean space. It is the L2-norm of the difference.
    
 - *Manhattan*: $||a-b||_1 = \displaystyle\sum_{i}|a_i - b_i|$
    
   - The distance between two points is the sum of the absolute differences of their Cartesian coordinates. It is the L1-norm of the difference.
        
 - *Cosine*: $1 - \frac{<x, y>}{||x||_2  ||y||_2} =  1 - \frac{\displaystyle\sum_{i}^{n} x_i y_i}{\sqrt{\displaystyle\sum_{i}^{n} x_i^2} {\sqrt{\displaystyle\sum_{i}^{n} y_i^2}}}$
    
   - Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. It contains the dot product scaled by the product of the Euclidean distances from the origin. 

    
    
    
- The ***linkage*** criterion determines the distance between sets of observations as a function of the distances between observations. In this analisys we focused on the main common criteria: ward, average, complete (also knows as MAX) and single (also knows as MIN).

*Ward*: Similarity of two clusters is based on the increase in squared error when two clusters are merged. The greater the SE value is the more the clusters are different.
    
*Complete*: Similarity of two clusters is based on the two least similar (most distant) points in the different clusters
<img src="https://imgur.com/DyLxthe.png" width="400">

*Single*: Similarity of two clusters is based on the two most similar (closest) points in the different clusters.
<img src="https://imgur.com/PGgU5Hh.png" width="400">
  
*Average*: Proximity of two clusters is the average of pairwise proximity between points in the two clusters.
<img src="https://imgur.com/q3TOARu.png" width="400">
    
In order to find the best parameters for our purpose we have performed a grid search as follows:

```python
from sklearn.cluster import AgglomerativeClustering

linkage_range = ['ward', 'complete', 'average', 'single']
affinity_range = ['euclidean', 'l2', 'manhattan', 'cosine']

best_silhouette = -1
for linkage in linkage_range:
    for affinity in affinity_range:
        if linkage == 'ward':
            affinity = 'euclidean'
            # With linkage ward only affinity = euclidean is allowed

        hierarchical = AgglomerativeClustering(n_clusters=len(label_mapping), affinity=affinity, linkage=linkage)
        
        hierarchical.fit_predict(X)
        
        labels = hierarchical.labels_ 
        
        silhouette = retrive_metrics_cluster(X, predicted_labels, affinity)
    
        if(silhouette > best_silhouette):
            best_silhouette = silhouette
            best_affinity = affinity
            best_linkage = linkage
        
        print("Hierarchical " + linkage + " " + affinity + " silhouette\t= %3.2f%%" %(silhouette*100))
```

### DBSCAN

Density-based spatial clustering of applications with noise (**DBSCAN**) is a density-based clustering algorithm: its cluster definition is based on the density-reachability. A point $q$ is directly reachable from another point $p$ if:
- their distance is less than a threshold ($\epsilon$) 
- $p$ is surrounded by a sufficient number of points

If these conditions are true, it is possible to consider $p$ and $q$ as two point of the same cluster.

Notations:
- ***$\epsilon$***: maximum distance between two points for them to be considered as in the same neighborhood.
- ***minPts***: minimum number of points required to form a cluster
- ***Core point***: a point which has more than a specified number of points (MinPts) within his neighborhood (intentified by a circumference of radius $\epsilon$)
- ***Border point***: a point which has fewer than MinPts within Eps, but is in the neighborhood of a core point
- ***Noise point***: a point that is not a core point or a border point

<img src="https://imgur.com/8CWUyyh.png" width="400">


This algorithm is an iterative one. It starts with a core point calculating its $\epsilon$-neighborhood. If it contains a sufficient number of points (greater or equal to minPts) a new cluster is created. If not, this point is classified as noise and it could become part of another cluster in the following interations. When a point is attached to one cluster, even the points in its $\epsilon$-neighborhood become part of the same cluster and the points in their $\epsilon$-neighborhood and so on. This process keeps going on until the cluster is completed. After it is completed, DBSCAN starts analysing another core point. The algorithm ends when it has analysed all the points in the dataset.

The quality of DBSCAN depends on the distance measure used. We have observed the best performances using *Euclidean distance*. However, especially for high-dimensional data, this is not enought due to the so-called "Curse of dimensionality", making it difficult to find an appropriate value for ε. This is the reason why the clusterization does not have been that good.
In the picture above we have considered the three main scenarios:
- Using `ε` = $3*10^8$ and `minPts` $= 1$ the Silhouette is negative which means a bad clusterization. This is confermed by the number of different clusters created (109 clusters, definitely too many).
- Using `ε` = $10^8$ and `minPts` $= 3$ the performances are quite acceptable but the clusterization is still a bad one, since the number of clusters created is two small (2 actual clusters and a noise one, definetly too few).
- Using `ε` = $9*10^8$ and `minPts` $= 1$ the performances are the best ones, since the number of clusters created is reasonable (17 clusters).


***Pros***:
 - The algorithm does not require to specify the number of clusters in the data a priori, as opposed to k-means.
 - The algorithm can find arbitrarily shaped clusters, not only globular ones. It can even find a cluster completely surrounded by a different cluster.
 - The algorithm has a notion of noises and is robust to outliers.

***Cons***:
- The algorithm is not entirely deterministic: border points that are reachable from more than one cluster can be part of either cluster, depending on the order the data are processed.
- The quality of DBSCAN depends on the distance measure used.
- The algorithm cannot cluster data sets well with large differences in densities, since the minPts-ε combination cannot then be chosen appropriately for all clusters.



### Comparison of Clustering Algorithms
---

The ***K-means*** algorithm is the simplest of the ones considered in this analysis, but it turned out to be a sufficiently accurate model, especially considering the high speed with which it converges. The ***Hierarchical*** algorithm, instead, turned out to be the worst one, both in terms of Silhouette and performances. This is reasonable because it has to calculate a lot of different distances matrixes and these are computionally challenging operations. Finally, the ***DBSCAN*** turned out to be a very versatile algorithm and, because of this, a precise *parameter tuning* was necessary to make it as efficient as possible. Once this step had been made, its performances were much better than those of the two previous algorithms analyzed. This is mainly due to the peculiarity of DBSCAN: it is able to cluster a dataset regardless of the form of the data which are contained in it. The Kmeans, instead, has difficulties to cluster non-globular datasets.
